```python
data=np.load("Etestdata2.npy",allow_pickle=True)[:100,4:]
print(data.shape)
ls=[]
for i in range(data.shape[0]):
    x1=data[i]
    x1[0]=0
    x1[-1]=0
    t1=np.linspace(0, 1, num=x1.shape[0], endpoint=True)
    plt.plot(t1,x1)
    ls+=[np.stack((x1,t1))]
plt.show()

plt.plot(t1,data.sum(axis=0)/1000)
plt.show()
```

```python
mean, aligned,warping,t = precise_karcher_mean(ls)
```

```python
for x in aligned:
    plt.plot(t,x)
    plt.plot(t,mean)
plt.show()

for x in warping:
    plt.plot(t,x)
plt.show()

plt.plot(t,mean)
plt.show()

t=center_warping(warping,t)
for x in aligned:
    plt.plot(t,x)
    plt.plot(t,mean)
plt.show()


for x in warping:
    plt.plot(t,x)
plt.show()

plt.plot(t,mean)
plt.show()
```
